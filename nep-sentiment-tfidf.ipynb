```python
import numpy as np
import pandas as pd
```

```python
df_train=pd.read_csv('/kaggle/input/second-train-test/df_train_2.csv')
df_test=pd.read_csv('/kaggle/input/second-train-test/df_test_2.csv')
```

```python
X_train = df_train['text']
X_test = df_test['text']
y_train=df_train['label']
y_test=df_test['label']
```

```python
import re
def text_clean(text):
    text=str(text)
    cleaned_text = re.sub(r'[^\w\s]', '', text)
    return cleaned_text
```

```python
#print(text_clean("Hello! This is @ sample text with some #special$characters."))
```

```python
X_train=X_train.apply(text_clean)
X_test=X_test.apply(text_clean)
```

```python
X_train.head(30)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()

X_train = tfidf.fit_transform(X_train)
X_test = tfidf.transform(X_test)
```

```python
X_train.shape
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,f1_score

log_model=LogisticRegression()
log_model.fit(X_train,y_train)
y_pred=log_model.predict(X_test)

log_f1=f1_score(y_test,y_pred,average='weighted')
print("Logistic_regression F1 Score:",log_f1)
```

```python
from sklearn.svm import SVC

svc_model = SVC()
svc_model.fit(X_train, y_train)
y_pred_svc = svc_model.predict(X_test)

svc_f1 = f1_score(y_test, y_pred_svc, average='weighted')

print("SVC F1 Score:", svc_f1)
```
